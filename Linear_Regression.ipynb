```python
features=["Day", "Confirmed"]
df3=df[features]
df3.head(31)
```

```python
%matplotlib inline
plt.xlabel("Day", fontsize=10)
plt.ylabel("Confirmed", fontsize=10)
plt.scatter(df3.head(31).Day,df3.head(31).Confirmed)
```

```python
RegressionObject = linear_model.LinearRegression()
RegressionObject.fit(df3.head(31).Day,df3.head(31).Confirmed)
```

```python
RegressionObject.predict([[32]])
```

```python
RegressionObject.predict([[25]])
```

```python
RegressionObject.score(df3.head(31)[["Day"]],df3.head(31).Confirmed)
```
